```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def future_value(PV, r, years):
    return PV*(1+r)**years
```

```python
for year in (3, 5, 10):
    for rate in (0.03, 0.05, 0.07):
        print("Future value of invesment of $1000 after " + str(year) + " years, with interest rate of " + str(rate) + " " + str(np.round(future_value(1000, rate, year), 2)) + ".")
```

```python
def forward_contract(FV, r, years):
    return FV*(1+r)**years
```

```python
for year in (1, 2, 10):
    for rate in (0.03, 0.05, 0.07):
        for future_value in (500, 1000, 1500):
            print("Forward price for spot price of " + str(future_value) + " for time period of " + str(year) + " years, with interest rate of " + str(rate) + " " + str(np.round(forward_contract(future_value, rate, year), 2)) + ".")
```

```python
#kupuj opcj do sprzedania za X, strike - to za ile moesz kupi

def long_call_option(price, asset_value, strike):
    if asset_value > strike:
        return asset_value - strike - price
    else:
        return - price 

#kupuj opcj do kupienia za X    
def long_put_option(price, asset_value, strike):
    if asset_value < strike:
        return strike - asset_value - price
    else:
        return - price 
    
```

```python
asset_value = np.arange(100)
plt.plot(asset_value, [long_call_option(10, asset_value[i], 50) for i in np.arange(len(asset_value))])
plt.title("Long call with spot price of \$50 and price of \$10")
```

```python
asset_value = np.arange(100)
plt.plot(asset_value, [long_put_option(10, asset_value[i], 50) for i in np.arange(len(asset_value))])
plt.title("Short call with spot price of \$50 and price of \$10")
```

```python
asset_value = np.arange(100)
plt.plot(asset_value, [long_call_option(5, asset_value[i], 30) for i in np.arange(len(asset_value))])
plt.title("Long call with spot price of \$30 and price of \$5")
```

```python
asset_value = np.arange(100)
plt.plot(asset_value, [-long_call_option(5, asset_value[i], 30) for i in np.arange(len(asset_value))])
plt.title("Short call with spot price of \$30 and price of \$5")
```

```python
plt.plot(np.arange(80), [long_call_option(10, asset_value[i], 20) - 2*long_call_option(10, asset_value[i], 40) + long_call_option(10, asset_value[i], 60) for i in np.arange(80)])
plt.title("Butterfly possition")
```

```python
plt.plot(np.arange(80), [long_call_option(10, asset_value[i], 20) - long_call_option(10, asset_value[i], 60) for i in np.arange(80)])
plt.title("Bull spread")
```

```python
plt.plot(np.arange(80), [long_call_option(10, asset_value[i], 60) - long_call_option(10, asset_value[i], 20) for i in np.arange(80)])
plt.title("Bear spread")
```

```python
plt.plot(np.arange(80), [long_call_option(10, asset_value[i], 40) + long_put_option(10, asset_value[i], 40) for i in np.arange(80)])
plt.title("Long straddle")
```

```python
plt.plot(np.arange(80), [-long_call_option(10, asset_value[i], 40) - long_put_option(10, asset_value[i], 40) for i in np.arange(80)])
plt.title("Short straddle")
```
